```python
df['location'] = df['img-link href'].apply(lambda x: x.split('/')[3])
df['city'] = df['location'].str.split('-').str[0]
df['district'] = df['location'].str.split('-').str[1]
df['neighborhood'] = df['location'].str.split('-').str[2:-1]
df['neighborhood'] = df['neighborhood'].apply(lambda x: '_'.join(x))
```

```python


def drop_columns(df, columns):
    return df.drop(columns=columns, errors='ignore')

df = drop_columns(df, ['location', 'img-link href'])
```

```python
replace_dict = {
    'Kot 2': '-2. Kat',
    'Kot 1': '-1. Kat',
    'Yüksek Giriş': '1. Kat',
    'Ara Kat': '3. Kat',
    'En Üst Kat': '5. Kat',
    'Bahçe Katı': '0. Kat',
    'Yarı Bodrum': '0. Kat',
    'Bodrum': '0. Kat',
    'Kot 3': '-3. Kat',
    'Çatı Katı': '5. Kat',
    'Zemin': '0. Kat',
    'Giriş Katı': '0. Kat',
    'Villa Katı': '0. Kat',
    '21 ve üzeri': '21. Kat',
    'Bodrum ve Zemin': '0. Kat',
    'Asma Kat': '1. Kat',
    'Tripleks': '0. Kat',
    'Teras Katı': '5. Kat',
    'nan': '2. Kat',
}

# Replace values using the dictionary
df['celly 4'] = df['celly 4'].replace(replace_dict)

# Replace any remaining 'nan' string or np.nan with '2. Kat'
df['celly 4'] = df['celly 4'].replace(['nan', np.nan], '2. Kat')

# Now extract the floor number and convert to int
df['floor'] = df['celly 4'].apply(lambda x: x.split('.')[0]).astype(int)
```
